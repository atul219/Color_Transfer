```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib qt
```

```python
source_img = cv2.imread('./source.jpg')
target_img = cv2.imread('./target.jpg')
```

```python
source_img = cv2.cvtColor(source_img , cv2.COLOR_BGR2RGB)
target_img = cv2.cvtColor(target_img , cv2.COLOR_BGR2RGB)
```

```python
fig , (axs1 , axs2) = plt.subplots(1,2 , figsize = (15,10))
axs1.set_title("Source Image")
axs1.imshow(source_img)
axs2.set_title("Target Image")
axs2.imshow(target_img)
```

```python
# convert image into LAB color space
source_lab = cv2.cvtColor(source_img , cv2.COLOR_RGB2LAB)
target_lab = cv2.cvtColor(target_img , cv2.COLOR_RGB2LAB)
```

```python
fig , (axs1 , axs2) = plt.subplots(1,2 , figsize = (15,10))
axs1.set_title("Source Lab Image")
axs1.imshow(source_lab)
axs2.set_title("Target Lab Image")
axs2.imshow(target_lab)
```

```python
# calculating mean and std for individual channels from source and target image
def image_parameters(L , A , B):
   # (L , A , B) = np.split(img)
    
    (L_mean , L_std) = (L.mean() , L.std())
    (A_mean , A_std) = (A.mean() , A.std())
    (B_mean , B_std) = (B.mean() , B.std())
    
    return (L_mean , L_std , A_mean , A_std , B_mean , B_std)

```

```python
def color_transfer(source , target):
    # convert image into LAB color space
    source_lab = cv2.cvtColor(source_img , cv2.COLOR_RGB2LAB)
    target_lab = cv2.cvtColor(target_img , cv2.COLOR_RGB2LAB)
    
    # split the channels of images
    (L_source , A_source , B_source) = cv2.split(source_lab)
    (L_target , A_target , B_target) = cv2.split(target_lab)
    
    # get the parameters of source and target images
    (L_source_mean , L_source_std , A_source_mean , A_source_std , B_source_mean , B_source_std) = image_parameters(L_source , A_source , B_source)
    (L_target_mean , L_target_std , A_target_mean , A_target_std , B_target_mean , B_target_std) = image_parameters(L_target , A_target , B_target)
    
    # now subtract every individual channel mean from data point
    # target image
    L_target = L_target - L_target_mean
    A_target = A_target - A_target_mean
    B_target = B_target - B_target_mean
    
    # now scale the data points by factor determined std
    L_target = (L_target_std / L_source_std) * L_target
    A_target = (A_target_std / A_source_std) * A_target
    B_target = (B_target_std / B_source_std) * B_target
    
    # now add source images mean to new target mean values
    L_new = L_target + L_source_mean
    A_new = A_target + A_source_mean
    B_new = B_target + B_source_mean
    
    # cliping the values
    L_new = np.clip(L_new , 0 , 255)
    A_new = np.clip(A_new , 0 , 255)
    B_new = np.clip(B_new , 0 , 255)
    
    transfer_img = cv2.merge([L_new , A_new , B_new])
    
    transfer_img = cv2.cvtColor(transfer_img.astype('uint8') , cv2.COLOR_LAB2RGB)
    
    return transfer_img
```

```python
transfer_img = color_transfer(source_img , target_img)
```
